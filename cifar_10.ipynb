Importing Libraries

```
import tensorflow as tf
import matplotlib.pyplot as plt
```

Loading Datasets from tf.keras.datasets and normalizing them for better accuracy

```
data = tf.keras.datasets.cifar10
(x_train,y_train),(x_test,y_test) = data.load_data() 
x_train, x_test = x_train/255.0, x_test/255.0
```

```
x_train.shape
```

Neural network model

```
cifar = tf.keras.models.Sequential()
## Layer 1
cifar.add(tf.keras.layers.Conv2D(32,(3,3),1,'same',input_shape=(32,32,3)))
cifar.add(tf.keras.layers.ReLU())
cifar.add(tf.keras.layers.BatchNormalization())
cifar.add(tf.keras.layers.Dropout(0.2))
cifar.add(tf.keras.layers.Conv2D(64,(3,3),1,'same'))
cifar.add(tf.keras.layers.ReLU())
cifar.add(tf.keras.layers.BatchNormalization())
cifar.add(tf.keras.layers.Conv2D(128,(3,3),1,'same'))
cifar.add(tf.keras.layers.ReLU())
cifar.add(tf.keras.layers.BatchNormalization())
cifar.add(tf.keras.layers.Dropout(0.2))
cifar.add(tf.keras.layers.AveragePooling2D((2,2),2,'same'))

## Layer 2
cifar.add(tf.keras.layers.Conv2D(128,(3,3),1,'same'))
cifar.add(tf.keras.layers.ReLU())
cifar.add(tf.keras.layers.BatchNormalization())
cifar.add(tf.keras.layers.Dropout(0.2))
cifar.add(tf.keras.layers.Conv2D(256,(3,3),1,'same'))
cifar.add(tf.keras.layers.ReLU())
cifar.add(tf.keras.layers.BatchNormalization())
cifar.add(tf.keras.layers.Dropout(0.2))
cifar.add(tf.keras.layers.Conv2D(256,(3,3),1,'same'))
cifar.add(tf.keras.layers.ReLU())
cifar.add(tf.keras.layers.BatchNormalization())
cifar.add(tf.keras.layers.Dropout(0.2))
cifar.add(tf.keras.layers.Conv2D(256,(3,3),1,'same'))
cifar.add(tf.keras.layers.ReLU())
cifar.add(tf.keras.layers.BatchNormalization())
cifar.add(tf.keras.layers.Dropout(0.2))
cifar.add(tf.keras.layers.AveragePooling2D((2,2),2,'same'))

## Layer 3

cifar.add(tf.keras.layers.Conv2D(512,(3,3),1,'same'))
cifar.add(tf.keras.layers.ReLU())
cifar.add(tf.keras.layers.BatchNormalization())
cifar.add(tf.keras.layers.Dropout(0.2))
cifar.add(tf.keras.layers.Conv2D(512,(3,3),1,'same'))
cifar.add(tf.keras.layers.ReLU())
cifar.add(tf.keras.layers.BatchNormalization())
cifar.add(tf.keras.layers.Dropout(0.2))
cifar.add(tf.keras.layers.AveragePooling2D((2,2),2,'same'))

## Layer 4
cifar.add(tf.keras.layers.Flatten())
cifar.add(tf.keras.layers.Dense(10))
cifar.add(tf.keras.layers.Activation(tf.keras.activations.softmax))
```

Model Summary

```
cifar.summary()
```

Losses

```
cifar.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(),metrics=['accuracy'])
cb_1 = tf.keras.callbacks.ModelCheckpoint(file_path_1,save_best_only=True,save_weights_only=True)
cb_2 = tf.keras.callbacks.ModelCheckpoint(file_path_2,save_weights_only=True)
cb_3 = tf.keras.callbacks.ModelCheckpoint(file_path_3,save_best_only=True,save_weights_only=True,monitor='val_accuracy')
```

Training

```
cifar_10=cifar.fit(x_train,y_train,batch_size=32,epochs=20,verbose=1,validation_split=0.2,validation_batch_size=32,callbacks=[cb_1,cb_2,cb_3])
```

```
cifar.load_weights(file_path_3)
```

```
cifar.evaluate(x_test, y_test)
```

```
epochs = range(35)
plt.plot(epochs,cifar_10.history['accuracy'],'b',label='Training_Acc')
plt.plot(epochs,cifar_10.history['val_accuracy'],'r',label = 'Val_Acc')
plt.title('Graph')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```
